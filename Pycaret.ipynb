<center><h2><span style="color:Blue">3. Modelos </span> </h2></center>

O Pycaret atende 6 modelos diferentes:

**Classification** - Modelo utilizado para classificar elementos em grupos, alguns exemplos são a previsão de padrão do cliente (sim ou não), rotatividade (ficar ou sair) e doenças (positivo ou negativo).

**Regression**  - Modelo utilizado para estimar as relações entre uma variável dependente e uma, ou mais, variáveis independentes. O objetivo é prever valores contínuos, como previsão de valores de vendas e previsão de temperatura.

**Clustering** - Modelo utilizado para agrupar um conjunto de objetos de forma que os objetos no mesmo grupo (também conhecido como cluster) sejam mais semelhantes entre si do que em outros grupos.

**Anomaly Detection** - Modelo utilizado para identificar itens raros, eventos ou observações que levantam suspeitas por diferir significativamente da maioria dos dados.

**Natural Language Processing** - Modelo utilizado para analisar dados de texto, criando modelos de tópico que podem encontrar estruturas semânticas ocultas em documentos.

**Association Rules Mining** - Modelo baseado em regras para descobrir relações interessantes entre variáveis em grandes bancos de dados. A intenção é identificar regras fortes usando medidas de interesse. Este modelo transforma automaticamente qualquer banco de dados transacional em uma forma que seja aceitável para o algoritmo apriori.

<center><h2><span style="color:Blue">3.1 Métricas </span> </h2></center>

- FN (Falso negativo): Modelo diz que a pessoa não vai sair e ela sai.

- FP (Falso positivo): Modelo diz que a pessoa vai sair e ela não sai.

- Recall: Quanto mais alto o recall menos FN gera.

- Precision: Quanto mais alto a precision menos FP gera.

- F1: Balanceia recall e precision.

- Acurácia: Indica uma performance geral do modelo. Dentre todas as classificações, quantas o modelo classificou corretamente.

<center><h2><span style="color:Blue">4. Escolhendo o Modelo </span> </h2></center>

**create_model**: Função treina e avalia o modelo passado como parâmetro.

```python
model = create_model('lr')
```

```python
# Apresenta todos os parâmetros que foram usados no modelo

print(model)
```

**tune_model**: Essa função ajusta os hiperparâmetros de um determinado modelo para melhorar a métrica.

```python
tuned_model = tune_model(model)
```

```python
print (tuned_model)
```

<center><h2><span style="color:Blue">5. Avaliando o Modelo </span> </h2></center>

**evaluate_model**: Exibe várias formas de visualização para dos resultados obtidos.

```python
evaluate_model(tuned_model)
```

```python
plot_model(tuned_model, plot = 'feature')
```

<center><h2><span style="color:Blue">6. Salvando o Modelo </span> </h2></center>

**save_model**: Esta função salva o pipeline de transformação e o objeto de modelo treinado no diretório de trabalho atual como um arquivo pickle para uso posterior.

```python
save_model(tuned_model,'modelo_salvo')
```
